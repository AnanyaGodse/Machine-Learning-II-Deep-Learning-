### Lab 2 - Stochastic & Batch Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0.5, 2.5])
Y = np.array([0.2, 0.9])
```

```python
def perceptron(x, w, b):
    return np.dot(x, w) + b
```

```python
def sigmoid(y_in):
    return 1/(1 + np.exp(-y_in))
```

```python
def stochastic_gradient_descent(X, Y):
    weights = []
    biases = []
    losses = []
    epochs = 100
    c = 0.1
    
    w = 0
    b = 0
    
    for i in range(epochs):
        for x, y in zip(X, Y):
            y_in = perceptron(x, w, b)
            y_hat = sigmoid(y_in)
            dw = c * (y - y_hat) * y_hat * (1 - y_hat) * x
            db = c * (y - y_hat) * y_hat * (1 - y_hat)
            w = w - dw
            b = b - db
        
        weights.append(w)
        biases.append(b) 
        loss = np.mean([(y - sigmoid(perceptron(x, w, b)))**2 for x, y in zip(X, Y)])
        losses.append(loss)
            
    return weights, biases, losses
         
```

```python
weights, biases, losses = stochastic_gradient_descent(X, Y)
```
